# Explore SFR diversions - Part 3

This example is a modified version of the Streamflow Routing (SFR) Package described in [Prudic et al., (2004)](https://doi.org/10.3133/ofr20041042). The problem has been modified by converting all of the SFR reaches to use rectangular channels.

<div>

</div>

We are going to run the original model and then evaluate the other diversion priority constraints (`cprior`) to better understand how they function. The original model uses the `UPTO` priority constraint for the diversion from reach 4 to reach 10.

The `UPTO` diversion priority constraint is defined as:

> `UPTO` – if Q$_{DS}$ in reach `IFNO` is $\ge$ to the specified diversion flow `DIVFLOW`, Q$_{DS}$ is reduced by `DIVFLOW`. If Q$_{DS}$ in reach `IFNO` is $\lt$ than `DIVFLOW`, `DIVFLOW` is set to Q$_{DS}$ and there will be no flow available for reaches connected to downstream end of reach `IFNO` for the diversion.

<div>

</div>

The same priority constraints are also used by the Mover (MVR) Package. See [Morway and others (2021)](https://doi.org/10.1111/gwat.13117) for more details on the MVR package.

## Model setup

Create the MODFLOW packages for standard packages.

```python
sim = flopy.mf6.MFSimulation(sim_name=name, sim_ws=ws)
flopy.mf6.ModflowTdis(
    sim, nper=nper, perioddata=tdis_ds, time_units=time_units
)
flopy.mf6.ModflowIms(
    sim,
    print_option="summary",
    linear_acceleration="bicgstab",
    outer_maximum=nouter,
    outer_dvclose=hclose,
    inner_maximum=ninner,
    inner_dvclose=hclose,
    rcloserecord=f"{rclose} strict",
)
gwf = flopy.mf6.ModflowGwf(sim, modelname=name, newtonoptions="newton")
flopy.mf6.ModflowGwfdis(
    gwf,
    length_units=length_units,
    nlay=nlay,
    nrow=nrow,
    ncol=ncol,
    delr=delr,
    delc=delc,
    idomain=idomain,
    top=top,
    botm=botm,
)

flopy.mf6.ModflowGwfnpf(
    gwf,
    icelltype=1,
    k=k11,
)

flopy.mf6.ModflowGwfsto(
    gwf,
    iconvert=1,
    sy=sy,
    ss=1e-5,
)
flopy.mf6.ModflowGwfic(gwf, strt=strt)

flopy.mf6.ModflowGwfghb(gwf, maxbound=len(ghb_spd), stress_period_data=ghb_spd)
flopy.mf6.ModflowGwfwel(
    gwf, maxbound=len(wel_spd[1]), stress_period_data=wel_spd
)
flopy.mf6.ModflowGwfrcha(gwf, recharge=recharge)
flopy.mf6.ModflowGwfevta(gwf, surface=surf, rate=evap_rate, depth=ext_depth)

flopy.mf6.ModflowGwfoc(
    gwf,
    printrecord=[("BUDGET", "ALL")],
);
```

### Create the SFR package

```python
sfr = flopy.mf6.ModflowGwfsfr(
    gwf,
    length_conversion=3.28081,
    nreaches=len(sfr_pakdata),
    packagedata=sfr_pakdata,
    connectiondata=sfr_conn,
    diversions=sfr_div,
    perioddata=sfr_spd,
    pname="sfr-1",
)
```

### Load the sfr observations

```python
# function to process the sfr observations and return a Panda dataframe
def get_sfr_obs():
    arr = gwf.sfr.output.obs().data.copy()
    arr = np.insert(arr, 0, 0.0, axis=0)
    for name in arr.dtype.names[1:]:
        arr[name][0] = float(arr[name][1])
    arr["totim"] /= 86400.0
    arr["R04_FLOW"] *= -1.0

    init_date = datetime.datetime(1954, 7, 1, 12, 0, 0)
    sim_datetime = []
    for dt in arr["totim"]:
        sim_datetime.append(init_date + datetime.timedelta(days=dt))

    return pd.DataFrame.from_records(arr, index=sim_datetime).drop(
        columns=["totim"]
    )
```

```python
# function to initialize a plot
def initialize_ts_plt(nrows=1):
    fig, ax = plt.subplots(
        ncols=1,
        nrows=nrows,
        sharex=True,
        figsize=(6.3, 3.15 * nrows),
        constrained_layout=True,
    )
    if nrows == 1:
        ax.set_ylabel("Discharge, m$^3$/s")
        ax.set_ylim(-1, 17)
    else:
        for a in ax:
            a.set_ylabel("Discharge, m$^3$/s")
            a.set_ylim(-1, 17)
    return fig, ax
```

```python
upto_df = get_sfr_obs()
upto_df
```

```python
with styles.USGSPlot():
    fig, axs = initialize_ts_plt(nrows=2)
    upto_df["R10_USFLOW"].plot(ax=axs[0], label="up to")
    styles.graph_legend(ax=axs[0])
    upto_df["R04_FLOW"].plot(ax=axs[1], label="reach 4 outflow", color="red")
    upto_df["R05_USFLOW"].plot(ax=axs[1], label="reach 5 inflow", color="cyan")
    styles.graph_legend(ax=axs[1])
```

## Change the diversion priority constraint to `EXCESS`

The `EXCESS` diversion priority constraint is defined as:

> CPRIOR = `EXCESS`, a diversion is made only if Q$_{DS}$ for reach IFNO exceeds the value of `DIVFLOW`. If this occurs, then the quantity of water diverted is the excess flow (Q$_{DS}$ - `DIVFLOW`) and Q$_{DS}$ from reach `IFNO` is set equal to `DIVFLOW`. This represents a flood-control type of diversion, as described by [Danskin and Hanson (2002)](https://pubs.usgs.gov/wri/wri024136/app.book.pdf).

<div>

</div>

\
To change the diversion priority constraint to `EXCESS` reset `sfr.diversions` to `[[3, 0, 9, "EXCESS"]]`.  You could also make these modifications manually in the SFR input file.

If using FloPy, the SFR package can be rewritten using `sfr.write()`. Run the simulation and plot the revised results.

```python
excess_df = get_sfr_obs()
excess_df
```

```python
with styles.USGSPlot():
    fig, ax = initialize_ts_plt()
    upto_df["R10_USFLOW"].plot(ax=ax, label="up to")
    excess_df["R10_USFLOW"].plot(ax=ax, label="excess")
    styles.graph_legend()
```

## Change the diversion priority constraint to `THRESHOLD`

The `THRESHOLD` diversion priority constraint is defined as:

> `CPRIOR` = `THRESHOLD`, then if Q$_{DS}$ in reach `IFNO` is $\lt$ the specified diversion flow `DIVFLOW`, no water is diverted from reach `IFNO`. If Q$_{DS}$ in reach `IFNO` is $\ge$ to DIVFLOW, DIVFLOW is diverted and Q_${DS}$ is set to the remainder (Q_${DS}$ - `DIVFLOW`)). This approach assumes that once flow in the stream is sufficiently low, diversions from the stream cease, and is the “priority” algorithm that originally was programmed into the STR1 Package ([Prudic, 1989](https://doi.org/10.3133/ofr88729)).

<div>

</div>

\
To change the diversion priority constraint to `THRESHOLD` reset `sfr.diversions` to `[[3, 0, 9, "THRESHOLD"]]`.  You could also make these modifications manually in the SFR input file. If using FloPy, rewrite the SFR package, run the simulation and plot the revised results.

```python
threshold_df = get_sfr_obs()
threshold_df
```

```python
with styles.USGSPlot():
    fig, ax = initialize_ts_plt()
    upto_df["R10_USFLOW"].plot(ax=ax, label="up to")
    excess_df["R10_USFLOW"].plot(ax=ax, label="excess")
    threshold_df["R10_USFLOW"].plot(ax=ax, label="threshold")
    styles.graph_legend()
```

## Change the diversion priority constraint to `FRACTION`

The `FRACTION` diversion priority constraint is the easiest to understand and is defined as:

> `CPRIOR` = `FRACTION`, then the amount of the diversion is computed as a fraction of the streamflow leaving reach `IFNO` (Q$_{DS}$); in this case, 0.0 $\le$ DIVFLOW $\le$ 1.0.

<div>

</div>

\
To change the diversion priority constraint to `FRACTION` reset `sfr.diversions` to `[[3, 0, 9, "FRACTION"]]`. You will also need to reset `sfr_spd[3][3]` from `15.0` to `0.5` and `sfr.perioddata` for the diversion using `sfr.perioddata.set_data(sfr_spd, key=0)`. You could also make these modifications manually in the SFR input file. If using FloPy, rewrite the SFR package, run the simulation and plot the revised results.

```python
fraction_df = get_sfr_obs()
fraction_df
```

```python
with styles.USGSPlot():
    fig, ax = initialize_ts_plt()
    upto_df["R10_USFLOW"].plot(ax=ax, label="up to")
    excess_df["R10_USFLOW"].plot(ax=ax, label="excess")
    threshold_df["R10_USFLOW"].plot(ax=ax, label="threshold")
    fraction_df["R10_USFLOW"].plot(ax=ax, label="fraction")
    styles.graph_legend()
```
